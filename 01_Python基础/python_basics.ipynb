## 2.Functions and Getting Help

```python
def add(a,b):
    return a+b

print(add(2,3))

def square(x):
    return x**2

print(square(9))
```

## 4. Lists

```python
courses=["python","pandas","ml"]
print(courses)
print(courses[1])
print(courses[2])
courses[2]="machine learning"
courses.append("deeplearning")
print(courses)
```

## 5. Loops and List Comprehensions

```python
for course in courses:
    print(course)

numbers=[1,2,3,4,5]
for number in numbers:
    print(number*2)

double_numbers =[x*2 for x in numbers]
print(double_numbers)
```
