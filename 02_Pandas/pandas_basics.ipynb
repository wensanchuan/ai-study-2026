## 2. Series and DataFrame

```python
import pandas as pd

data = {
    "name": ["A", "B", "C", "D"],
    "score": [80, 92, 76, 88],
    "group": ["A", "B", "A", "B"]
}
data ={
    'name':['A', 'B', 'C', 'D'],
    'score':[80, 92, 76, 88],
    'group':['A', 'B', 'A', 'B']
    }
df=pd.DataFrame(data)
df = pd.DataFrame(data)
print(df)
```

## 3. Select Columns
loc与iloc的区别：iloc 使用 Python 标准库的索引方案，其中包含范围的第一个元素，排除最后一个元素。因此， 0:10 将选择条目 0,...,9 。而 loc 则采用包含索引的方式。因此， 0:10 将选择条目 0,...,10

```python
print(df['score'])
print(df.loc[:, 'score'])
```

## 4. Filter Rows

```python
print(df[df['score'] >= 85])
print(df[df['group'] == 'A'])
print(df[(df['group'] == 'A') & (df['score'] >= 85)])
print(df[df['group'] == 'A']['score'])
print(df[df['group'] == 'A'].loc[:, 'score'])
```

## 8. Summary Functions and Maps

```python
import pandas as pd
data4 = {
    "student": ["Tom", "Jerry", "Lucy", "Lily", "Tom"],
    "age": [20, 21, 22, 20, 20],
    "score": [85, 70, 95, 88, 90],
    "group": ["A", "B", "A", "B", "A"]
}
df4 = pd.DataFrame(data4)
print(df4)
print(df4.duplicated()) 
print(df4.describe())
print(df4.mean(numeric_only=True))
print(df4.groupby("group")["age"].mean())
print(df4['score'].value_counts())
print(df4['score'].mean())
print(df4['score'].describe())
print(df4["score"].unique())
print(df4["score"].value_counts())
```

My Exercises - Day 8

```python
print(df4["score"].mean())
print(df4["score"].max())
print(df4["score"].min())

print(df4["group"].unique())
print(df4["group"].value_counts())

df4["score_plus_10"] = df4["score"].map(lambda x: x + 10)
print(df4)

df4["passed"] = df4["score"] >= 80
print(df4)

def category_func(row):
    if row["score"] >= 90:
        return "A"
    elif row["score"] >= 80:
        return "B"
    else:
        return "C"

df4["category"] = df4.apply(category_func, axis=1)
print(df4)
```

## 10. Grouping and Sorting

```python
import pandas as pd

data5 = {
    "student": ["Tom", "Jerry", "Lucy", "Lily", "Tom", "Lucy"],
    "group": ["A", "B", "A", "B", "A", "A"],
    "score": [85, 70, 95, 88, 90, 78],
    "age": [20, 21, 22, 20, 20, 22]
}

df5 = pd.DataFrame(data5)
print(df5)
```

```python
print(df5.groupby("group").group.size())
```

```python
print(df5.sort_values(by=["group", "score"], ascending=[True, False]))
```

## My Exercises - Day 9

```python
print(df5.groupby("group").group.count())
print(df5.groupby("group").score.mean())
print(df5.groupby("group").score.max())
print(df5.groupby("group").age.mean())

print(df5.sort_values(by="score", ascending=False))
print(df5.sort_values(by=["group", "age"]))
```

## 14. Renaming and Combining

```python
import pandas as pd

df_a=pd.DataFrame({
    'student':['Tom','Jerry'],
    'score':[85,70]
    })

df_b=pd.DataFrame({
    'student':['Lucy','Lily'],
    'score':[95,88]
    })

print(df_a.rename(columns={'student':'name'}))
df_all=pd.concat([df_a,df_b])
print(df_all)
```

## 15. Pandas Mini Practice

```python
import pandas as pd
import numpy as np

data = {
    "student": ["Tom", "Jerry", "Lucy", "Lily", None],
    "group": ["A", "B", "A", "B", "A"],
    "score": [85, 70, 95, np.nan, 88],
    "age": [20, 21, 22, 20, np.nan]
}

df = pd.DataFrame(data)

print("原始表：")
print(df)

print("\n选取 score 列：")
print(df["score"])

print("\n筛选 score >= 80 的行：")
print(df[df["score"] >= 80])

print("\n按 group 分组后的平均分：")
print(df.groupby("group")["score"].mean())

print("\n按 score 降序排序：")
print(df.sort_values(by="score", ascending=False))

df["score_filled"] = df["score"].fillna(0)
df["age_filled"] = df["age"].fillna(0)
df["passed"] = df["score_filled"] >= 80

print("\n补缺失值并新增 passed 列后：")
print(df)

df2 = df.rename(columns={"student": "name"})
print("\n改列名后：")
print(df2)
```
